### Subsets

Given a set of distinct integers, nums, return all possible subsets (the power set).

Note: The solution set must not contain duplicate subsets.
##### Example:
Input: nums = [1,2,3]

Output:

[

[3],

[1],

[2],

[1,2,3],

[1,3],

[2,3],

[1,2],

[]

]

```python
def subsets(self, nums: list) -> list:
    new = [[]]
        
    for i in nums:
        size = len(new)
        for j in range(size):
            new += [new[j] + [i]]
        
    return new
```
