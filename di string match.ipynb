### DI String Match

Given a string S that only contains "I" (increase) or "D" (decrease), let N = S.length.

Return any permutation A of [0, 1, ..., N] such that for all i = 0, ..., N-1:

If S[i] == "I", then A[i] < A[i+1]

If S[i] == "D", then A[i] > A[i+1] 
##### Example 1:
Input: "IDID"

Output: [0,4,1,3,2]
##### Example 2:
Input: "III"

Output: [0,1,2,3]
##### Example 3:
Input: "DDI"

Output: [3,2,0,1]

```python
def diStringMatch(self, S: str) -> list:
    c, d = 0, len(S)
    s = list(S)
        
    for i in range(len(s)):
        if s[i] == 'I':
            s[i] = c
            c += 1

        else:
                s[i] = d
                d -= 1
                
    e = c
        
    if len(set(S)) == 2:
        return s + [e]
        
    if len(set(S)) == 1 and s[0] == 0:
        return s + [len(S)]
        
    else:
        return s + [0]
```
