### Monotonic Array

An array is monotonic if it is either monotone increasing or monotone decreasing.

An array A is monotone increasing if for all i <= j, A[i] <= A[j].  An array A is monotone decreasing if for all i <= j, A[i] >= A[j].

Return true if and only if the given array A is monotonic.
##### Example 1:

Input: [1,2,2,3]

Output: true
##### Example 2:
Input: [6,5,4,4]

Output: true
##### Example 3:
Input: [1,3,2]

Output: false
##### Example 4:
Input: [1,2,4,5]

Output: true
##### Example 5:
Input: [1,1,1]

Output: true

```python
def isMonotonic(self, A: list) -> bool:
    if len(A) <= 2:
        return True
        
    if A[0] == A[-1]:
        if any(A[i] != A[0] for i in range(1, len(A))):
            return False
        else:
            return True
            
    elif A[0] < A[-1]:
        if any(A[i + 1] < A[i] for i in range(len(A) - 1)):
            return False
        else:
            return True
        
    else:
        if any(A[i + 1] > A[i] for i in range(len(A) - 1)):
            return False
        else:
            return True
```
