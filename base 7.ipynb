### Base 7

Given an integer, return its base 7 string representation.
##### Example 1:
Input: 100

Output: "202"
##### Example 2:
Input: -7

Output: "-10"

### Brief analysis
use while function and update the target each time

```python
def convertToBase7(num: int) -> str:
    n = abs(num)
    new = ''
    while n // 7 > 0:
        new += str(n % 7)
        n = n // 7
    new += str(n)
    
    return new[::-1] if num >= 0 else '-' + new[::-1]
```
