### Complex Number Multiplication

Given two strings representing two complex numbers.

You need to return a string representing their multiplication. Note i2 = -1 according to the definition.
##### Example 1:
Input: "1+1i", "1+1i"

Output: "0+2i"

Explanation: (1 + i) * (1 + i) = 1 + i2 + 2 * i = 2i, and you need convert it to the form of 0+2i.
##### Example 2:
Input: "1+-1i", "1+-1i"

Output: "0+-2i"

Explanation: (1 - i) * (1 - i) = 1 + i2 - 2 * i = -2i, and you need convert it to the form of 0+-2i.

Note:

The input strings will not have extra blank.

The input strings will be given in the form of a+bi, where the integer a and b will both belong to the range of [-100, 100]. And the output should be also in this form.

```python
def complexNumberMultiply(self, a: str, b: str) -> str:
    ind1, ind2 = a.index('+'), b.index('+')
        
        
    if a[ind1 + 1] == '-' and b[ind2 + 1] == '-':
        x = int(a[:ind1]) * int(b[:ind2]) - (int(a[ind1 + 2: -1]) * int(b[ind2 + 2: -1]))  
        y = int(a[:ind1]) * int(b[ind2 + 2: -1]) + int(b[:ind2]) * int(a[ind1 + 2: -1])
        if y == 0:
            return str(x) + '+0i'
        return str(x) + '+-' + str(y) + 'i' if y > 0 else str(x) + '+' + str(-y) + 'i'
        
    if a[ind1 + 1] == '-' and b[ind2 + 1] != '-':
        x = int(a[:ind1]) * int(b[:ind2]) + (int(a[ind1 + 2: -1]) * int(b[ind2 + 1: -1]))  
        y = int(a[:ind1]) * int(b[ind2 + 1: -1]) - int(b[:ind2]) * int(a[ind1 + 2: -1])
            
        return str(x) + '+' + str(y) + 'i'
        
    if a[ind1 + 1] != '-' and b[ind2 + 1] == '-':
        x = int(a[:ind1]) * int(b[:ind2]) + (int(a[ind1 + 1: -1]) * int(b[ind2 + 2: -1]))  
        y = int(b[:ind2]) * int(a[ind1 + 1: -1]) - int(a[:ind1]) * int(b[ind2 + 2: -1]) 
            
        return str(x) + '+' + str(y) + 'i'
           
            
    else:
        x = int(a[:ind1]) * int(b[:ind2]) - (int(a[ind1 + 1: -1]) * int(b[ind2 + 1: -1]))  
        y = int(a[:ind1]) * int(b[ind2 + 1: -1]) + int(b[:ind2]) * int(a[ind1 + 1: -1])
            
        return str(x) + '+' + str(y) + 'i'
```
