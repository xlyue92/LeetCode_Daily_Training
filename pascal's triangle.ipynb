### Pascal's Triangle

Given a non-negative integer numRows, generate the first numRows of Pascal's triangle.

Example:

Input: 5
Output:
[
     [1],
     
    [1,1],
    
   [1,2,1],
   
  [1,3,3,1],
  
 [1,4,6,4,1]
 
]

```python
def generate(self, numRows: int) -> list:
    res = []
        
    for i in range(numRows):
        res += [[]]
        for j in range(i + 1):
            if j in (0, i):
                res[i] += [1]
            else:
                res[i] += [res[i - 1][j - 1] + res[i - 1][j]]
        
    return res
```
