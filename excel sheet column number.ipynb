### Excel Sheet Column Number

Given a column title as appear in an Excel sheet, return its corresponding column number.

For example:

    A -> 1
    B -> 2
    C -> 3
    ...
    Z -> 26
    AA -> 27
    AB -> 28 
    ...
##### Example 1:
Input: "A"

Output: 1
##### Example 2:
Input: "AB"

Output: 28
###### Example 3:
Input: "ZY"

Output: 701

### Brief analysis
ord

```python
def titleToNumber(s: str) -> int:
    new = 0
    for i in s:
        new = 26 * new + ord(i) - ord('A') + 1
        
    return new
```
