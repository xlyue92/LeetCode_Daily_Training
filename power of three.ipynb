### Power of Three

Given an integer, write a function to determine if it is a power of three.
##### Example 1:
Input: 27

Output: true
##### Example 2:
Input: 0

Output: false
##### Example 3:
Input: 9

Output: true
##### Example 4:
Input: 45

Output: false

#### Brief analysis
similar to power of two

```python
def isPowerOfThree(n: int) -> bool:
        return 4052555153018976267 % n == 0 if n > 0 else False
```
