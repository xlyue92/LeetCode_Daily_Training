### Power of two

Given an integer, write a function to determine if it is a power of two.
##### Example 1:
Input: $1$

Output: true 

Explanation: $2^0 = 1$

##### Example 2:
Input: $16$

Output: true

Explanation: $2^4 = 16$

##### Example 3:
Input: $218$

Output: false

```python
def isPowerOfTwo(n: int) -> bool:
    if n > 0 and 2147483648 % n == 0:
        return True
    else:
        return False
```
