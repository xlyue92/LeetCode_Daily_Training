### Combinations

Given two integers n and k, return all possible combinations of k numbers out of 1 ... n.
##### Example:
Input: n = 4, k = 2

Output:
[
  [2,4],
  [3,4],
  [2,3],
  [1,2],
  [1,3],
  [1,4],
]

```python
def combine(self, n: int, k: int) -> list:
    nums = list(range(1, n + 1))
    new, res = [[]], []
        
    for i in nums:
        size = len(new)
        for j in range(size):
            new += [new[j] + [i]]
        
    for n in new:
        if len(n) == k:
            res += [n]
                
    return res
```
