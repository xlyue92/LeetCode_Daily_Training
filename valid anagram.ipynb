### Valid Anagram

Given two strings s and t , write a function to determine if t is an anagram of s.
##### Example 1:
Input: s = "anagram", t = "nagaram"

Output: true
##### Example 2:
Input: s = "rat", t = "car"

Output: false

```python
def isAnagram(self, s: str, t: str) -> bool:
    from collections import Counter as ct
        
    return ct(s) == ct(t)
```
