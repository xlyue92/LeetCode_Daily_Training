### Set Matrix Zeroes

Given a m x n matrix, if an element is 0, set its entire row and column to 0. Do it in-place.
##### Example 1:
Input: 
[
  [1,1,1],
  [1,0,1],
  [1,1,1]
]

Output: 
[
  [1,0,1],
  [0,0,0],
  [1,0,1]
]
##### Example 2:
Input: 
[
  [0,1,2,0],
  [3,4,5,2],
  [1,3,1,5]
]

Output: 
[
  [0,0,0,0],
  [0,4,5,0],
  [0,3,1,0]
]

```python
def setZeroes(self, matrix: list) -> None:
    """
    Do not return anything, modify matrix in-place instead.
    """
    r, c = [], []
    for i, p in enumerate(matrix):
        for j, q in enumerate(p):
            if q == 0:
                r += [i]
                c += [j]
        
    for k in matrix:
        for m in c:
            k[m] = 0
    for k in r:
        matrix[k] = [0 for x in range(len(matrix[0]))]
```
