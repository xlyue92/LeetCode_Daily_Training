### Hamming Distance

The Hamming distance between two integers is the number of positions at which the corresponding bits are different.

Given two integers x and y, calculate the Hamming distance.

Note:

0 ≤ x, y < 231.
##### Example:
Input: x = 1, y = 4

Output: 2

Explanation:

1   (0 0 0 1)

4   (0 1 0 0)
       
     ↑   ↑

The above arrows point to positions where the corresponding bits are different.

```python
def hammingDistance(self, x: int, y: int) -> int:
    if x == y:
        return 0
        
    m, n = '', ''
    while x // 2 > 0:
        m += str(x % 2)
        x = x // 2
    m += str(x)
    m = m[::-1]
        
    while y // 2 > 0:
        n += str(y % 2)
        y = y // 2
    n += str(y)
    n = n[::-1]
        
    c = 0
    if len(m) == len(n):
        for i, j in zip(m, n):
            if i != j:
                c += 1

    elif len(m) < len(n):
        m = '0' * (len(n) - len(m)) + m
        for i, j in zip(m, n):
            if i != j:
                c +=1
                
    else:
        n = '0' * (len(m) - len(n)) + n
        for i, j in zip(m, n):
            if i != j:
                c +=1
     
    return c
```
