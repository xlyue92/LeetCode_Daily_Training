### Valid Mountain Array

Given an array A of integers, return true if and only if it is a valid mountain array.

Recall that A is a mountain array if and only if:

A.length >= 3

There exists some i with 0 < i < A.length - 1 such that:

A[0] < A[1] < ... A[i-1] < A[i]

A[i] > A[i+1] > ... > A[A.length - 1]
##### Example 1:
Input: [2,1]

Output: false
##### Example 2:
Input: [3,5,5]

Output: false
##### Example 3:
Input: [0,3,2,1]

Output: true

```python
def validMountainArray(self, A: list) -> bool:
    if len(A) < 3:
        return False
        
    if len(A) == 3:
        return True if A[1] > A[0] and A[1] > A[2] else False
        
    new = [j - i for i, j in zip(A[:-1], A[1:])]
    if 0 in new:
        return False
        
    for k in range(len(new)):
        if new[k] < 0:
            ind = k
            break
        
    return False if any(n > 0 for n in new[k:]) or ind == 0 else True
```
